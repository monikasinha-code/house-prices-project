```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import RobustScaler
from sklearn.feature_selection import VarianceThreshold
from sklearn.linear_model import Ridge
```

```python
# 1. Load the dataset
df = pd.read_csv("final_model_ready_data.csv")
```

```python
# 2. Prepare features and target
target_col = "SalePrice"
id_col = "Id"

y = df[target_col]
feature_data = df.drop(columns=[target_col, id_col], errors="ignore")

# Keep only numeric and boolean columns, convert boolean to 0/1
X = feature_data.select_dtypes(include=["number", "bool"]).copy()
bool_cols = X.select_dtypes(include=["bool"]).columns
X[bool_cols] = X[bool_cols].astype(int)
```

```python
# 3. Data Sanitization
# Replace inf values and fill NaNs
X = X.replace([np.inf, -np.inf], np.nan).fillna(0)

# 4. Train/Test Split
RANDOM_STATE = 42
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=RANDOM_STATE
)
```

```python
# 5. Enhanced Pipeline
# Added VarianceThreshold to remove features with zero or near-zero variance
log_ridge_pipeline = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="median")),
    ("variance_filter", VarianceThreshold(threshold=0)), 
    ("scaler", RobustScaler()), 
    ("ridge", Ridge(random_state=RANDOM_STATE))
])
```

```python
# 6. Grid Search with higher alpha range to ensure numerical stability
# Using a higher start value for alpha to avoid matrix singularity
alpha_values = np.logspace(1, 5, 20) 

log_ridge_grid = GridSearchCV(
    estimator=log_ridge_pipeline,
    param_grid={"ridge__alpha": alpha_values},
    cv=5,
    scoring="neg_mean_squared_error",
    n_jobs=1
)
```

```python
# 7. Fit and Predict
log_ridge_grid.fit(X_train, np.log1p(y_train))

# Best model extracted from GridSearch
best_log_ridge = log_ridge_grid.best_estimator_

# Transform predictions back to original price scale
log_ridge_pred = np.expm1(best_log_ridge.predict(X_test))
log_ridge_pred = np.maximum(log_ridge_pred, 0)
```

```python
# 8. Output
rmse = np.sqrt(((y_test - log_ridge_pred) ** 2).mean())
print(f"Best Alpha: {log_ridge_grid.best_params_['ridge__alpha']:,.2f}")
print(f"Test RMSE with Log Transformation: {rmse:,.2f}")
```
