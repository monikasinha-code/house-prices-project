```python
import pandas as pd

# Load your finished files
df_num = pd.read_csv('numerical_engineered_data.csv')
df_cat = pd.read_csv('categorical_engineered_data.csv')
```

```python
# Identify common columns to avoid
cols_to_use = df_cat.columns.difference(df_num.columns).tolist()
cols_to_use.append('Id') # Make sure we keep the ID for the merge

# Merge using only the unique columns from the second dataframe
final_df = pd.merge(df_num, df_cat[cols_to_use], on='Id')
```

```python
# Save the final, unified dataset
final_df.to_csv('final_model_ready_data.csv', index=False)
print("Dataset successfully merged and ready for Phase 3!")
```
